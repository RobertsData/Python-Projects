```python
# Covid-19 Cases By country

import pandas as pd
from matplotlib import pyplot as plt, rcParams

# Source: Johns Hopkins Github Repository

drl = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
df = pd.read_csv(drl)
df = df.drop(['Province/State', 'Lat', 'Long'], axis=1) # Drop Extra Columns
df = df.groupby("Country/Region").sum() # Group by country, creating a sum of cases for each country
df = df.transpose()
df.index = pd.to_datetime(df.index) # Convert to DatetimeIndex

start_date = '2020-03-05' # Select Start Date of Graph
df = df[start_date:]

rcParams['font.size']=12 # Adjust Graph Font Size
countries = ['US','Italy', 'France', 'Spain', 'United Kingdom'] # Select countries to plot
current_figs = []

# Plot Graph of New Case Curves by Country

plt.figure(figsize=(12,8))
for c in countries:
    plt.plot(df[c])
    current_figs.append(max(df[c]))
plt.title('Total Confirmed Cases')
plt.legend(countries)
plt.xticks(rotation=45)
plt.show()

# New Case Counts

country_figures = {countries[i]: current_figs[i] for i in range(len(countries))}
print(country_figures)

# Number of New Cases per Day

data = pd.read_csv(drl) # reload original table
data = data.drop(['Lat', 'Long'], axis=1)
data = data.groupby("Country/Region").sum()
data=data.transpose()

country = 'US' # Choose country to plot

data = df[country]
data = pd.DataFrame(data)
data['diff'] = data.diff()

plt.figure(figsize=(12,8))
plt.plot(data['diff'])
plt.fill_between(data.index, data['diff'], alpha=0.3)
plt.xticks(rotation=45)
plt.show()
```

```python
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv"
dt = pd.read_csv(deaths_url)
dt = dt.drop(['Population','UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2','Province_State', 'Lat', 'Long_'], axis=1) # Drop Extra Columns
dt = dt.groupby("Country_Region").sum() # Group by country, creating a sum of cases for each country
dt = dt.transpose()
dt.index = pd.to_datetime(dt.index) # Convert to DatetimeIndex

start_date = '2020-03-01' # Select Start Date of Graph
dt = df[start_date:]
dt =dt['US']


dt = pd.DataFrame(dt)
dt['diff'] = dt.US.diff()
print(f'There are {dt.US.max()} total deaths')
print(dt.tail())
plt.figure(figsize=(12,8))
plt.plot(dt['diff'])
plt.fill_between(dt.index, dt['diff'], alpha=0.3)
plt.xticks(rotation=45)
plt.show()
print(dt.tail())
```
