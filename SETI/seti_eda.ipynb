```python
def show_cadence(filename: str, label: int) -> None:
    plt.figure(figsize=(16, 10))
    arr = np.load(filename)
    for i in range(6):
        plt.subplot(6, 1, i + 1)
        if i == 0:
            plt.title(f':File ID: {os.path.basename(filename)} TARGET: {label}', fontsize=18)
        plt.imshow(arr[i].astype(float), interpolation='nearest', aspect='auto')
        plt.text(5, 100, ["ON", "OFF"][i % 2], bbox={'facecolor': 'white'})
        plt.xticks([])
    plt.show()
```

```python
tmp_filepath = os.listdir('sample_data/')[0]

show_cadence(filename = 'sample_data/' + tmp_filepath, label = 0)
```
