```python
df = pd.read_csv('paris_restaurants.csv')
restaurants = geopandas.GeoDataFrame(df, geometry=geopandas.points_from_xy(df.x, df.y))
ax = restaurants.plot(markersize=1)
contextily.add_basemap(ax)
plt.show()
```

```python
belgium = cr[cr['ADMIN']=='Belgium']['geometry'].squeeze()
uk = cr[cr['ADMIN']=='United Kingdom']['geometry'].squeeze()
ne = cr[cr['ADMIN']=='Netherlands']['geometry'].squeeze()
fr = cr[cr['ADMIN']=='France']['geometry'].squeeze()

cities= geopandas.read_file(geopandas.datasets.get_path(geopandas.datasets.available[0]))
brussels = cities[cities.name=='Brussels'].iloc[0]['geometry']

nw_europe = geopandas.GeoSeries([belgium, ne, uk])
nw_europe.plot()

plt.xlim(-10, 10)
plt.ylim(45, 65)
```

```python
# contains, touches, within
print(belgium.contains(brussels))
print(ne.contains(brussels))
print(brussels.within(belgium))
print(belgium.touches(ne))
```

```python
# can also use shapely to create points from scratch
from shapely.geometry import Point
from pyproj import Proj
nys = Proj(init='EPSG:32117')

# Need better way to do this
bruges = Point(nys( 51.2029, 3.2248 ))
print(bruges.within(belgium))
```

```python
# can filter for cities within other polygons
cities[cities.within(uk)]
```

```python
chile = cr[cr['ADMIN']=='Chile']
cities.within(chile)
```

```python
# sjoin to join field points with coverage of each of the major carriers
# distance to calculate distance to every cell tower

# conver one geodataframe crs to another using to_crs()
cities.to_crs(cr.crs)
geopandas.sjoin(cities, cr[['ADMIN', 'geometry']], op='within')
```

```python
# coordinates WGS84, another is EPSG:4326
print(cities.crs)
print(cr.crs)
```

```python
# Create a point from scratch
eiffel_tower = Point( 2.2945, 48.8584)

# Put the point in a GeoSeries with the correct CRS
s_eiffel_tower = geopandas.GeoSeries([eiffel_tower], crs={'init': 'EPSG:4326'})

# Convert to other CRS
s_eiffel_tower_projected = s_eiffel_tower.to_crs(epsg=2154)

# Print the projected point
print(s_eiffel_tower_projected.within(fr))
print(s_eiffel_tower_projected)
```

```python
districts.contains(s_eiffel_tower_projected)
```
